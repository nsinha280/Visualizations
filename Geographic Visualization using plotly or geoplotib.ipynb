```python
import plotly.plotly as py
import pandas as pd

df1 = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_world_gdp_with_codes.csv')
df = data.merge(df1,left_on= 'Country', right_on='COUNTRY')
df.head()
```

```python
data = [ dict(
        type = 'choropleth',
        locations = df['CODE'],
        z = df['Happiness Score'],
        text = df['COUNTRY'] 
             ,
        colorscale = [[0,"rgb(5, 10, 172)"],[0.35,"rgb(40, 60, 190)"],[0.5,"rgb(70, 100, 245)"],\
            [0.6,"rgb(90, 120, 245)"],[0.7,"rgb(106, 137, 247)"],[1,"rgb(220, 220, 220)"]],
        autocolorscale = False,
        reversescale = True,
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
        colorbar = dict(
            
            title = 'Happiness Score'),
      ) ]

layout = dict(
    title = 'World Happiness score',
    geo = dict(
        showframe = False,
        showcoastlines = False,
        projection = dict(
            type = 'Mercator'
        )
    )
)

fig = dict( data=data, layout=layout )
#py.iplot( fig, validate=False, filename='d3-world-map' )
iplot(fig)
```

The above world map shows the of happiness score for each country.

```python
codes = pd.read_csv('codes.csv', header=None)
data= pd.read_csv('2015.csv')
df = data.merge(codes,left_on= 'Country', right_on=3)
df['text'] = df['Country'] +'<br>'+ 'Happiness Rank : ' + [str(x) for x in df['Happiness Rank']] +'<br>' + 'Happiness Score :' +'<br>'+ [str(x) for x in df['Happiness Score']]+ '<br>'+ 'Economy (GDP per Capita) : ' + [str(x) for x in df['Economy (GDP per Capita)']]+ '<br>' + 'Family : ' + [str(x) for x in df['Family']] + '<br>' + 'Health (Life Expectancy) : ' + [str(x) for x in df['Health (Life Expectancy)']] + '<br>' + 'Freedom : ' + [str(x) for x in df['Freedom']] 
scl = [ [0,"#d7191c"],[0.25,"#fdae61"], [0.5,"#ffffbf"],[0.75,"#abd9e9"], [1,"#2c7bb6"] ]
data = [ dict(
        type = 'scattergeo',
        locationmode = 'country_names',
        lon = df[2],
        lat = df[1],
        text = df['text'],
        mode = 'markers',
        marker = dict(
            size = 8,
            opacity = 0.8,
            reversescale = True,
            autocolorscale = False,
            symbol = 'circle-dot',
            line = dict(
                width=1,
                color='rgba(102, 102, 102)'
            ),
            colorscale = scl,
            cmin = 0,
            color = df['Happiness Rank'],
            cmax = df['Happiness Rank'].max(),
            colorbar=dict(
                title="Happiness Rank"
            )
        ))]

layout = dict(
        title = 'Happiness Rank<br>(Hover for country names)',
        
        geo = dict(
            scope='world',
            projection=dict(),
            showland = True,
            landcolor = "rgb(250, 250, 250)",
            subunitcolor = "rgb(217, 217, 217)",
            countrycolor = "rgb(217, 217, 217)",
            countrywidth = 0.5,
            subunitwidth = 0.5
        ),
    )

fig = dict( data=data, layout=layout )
iplot(fig)
```

The above world map shows the variation of happiness score across different countries. For most of the countries, the economic stability which is the GDP per capita contributes the most to the happiness score. The countries like USA, Canada which are rich have more happiness score and countries like those of Africa which are poor have low happiness score. The second most important factor is family which also becomes the first contributor for some countries like Costa Rica. The next important factors are health which impact the happiness score. The better the health facilities, more happiness score will be. Some countries like Rwanda has health as the major contributor in happiness score.
The next important factor is freedom which also contributes to happiness score. Some countries like Togo where the freedom for people is less has freedom as the major contributor. The other factors which contribute to happiness score are trust and generosity. 

Different countries have different situations and the major factors contributing to their happiness score also change.
